```r
library(abcrf)
# help(abcrf)
```

```r
s = seq(-.9,.9,.01)
nx=20
ny=10
nn=nx*ny*2
Data.healthy=data.frame(matrix(ncol = nn, nrow = 1))
n=0
for(i in sapply(s,function(x)paste("./diagnoses/s=", format(x, nsmall = 2),",diam_narrow=0.000000.csv",sep=""))){
    n=n+1
    x=read.csv(i,header=FALSE)
    stopifnot(dim(Data.healthy)[2]==length(c(data.matrix(x))))
    Data.healthy[n,]=c(data.matrix(x))
}
Data.sick=data.frame(matrix(ncol = nn, nrow = 1))
n=0
for(i in sapply(s,function(x)paste("./diagnoses/s=", format(x, nsmall = 2),",diam_narrow=0.030000.csv",sep=""))){
    n=n+1
    x=read.csv(i,header=FALSE)
    stopifnot(dim(Data.sick)[2]==length(c(data.matrix(x))))
    Data.sick[n,]=c(data.matrix(x))
}
```

```r
Data.sick$healthy=0
Data.healthy$healthy=1

Data.all=rbind(Data.healthy,Data.sick)
Data.all$healthy=factor(Data.all$healthy)            
```

```r
model.rf3 <- abcrf(healthy~., data = Data.all, ntree=100,lda=FALSE)
model.rf3
```

```r
model.rf1 <- abcrf(healthy~., data = Data.all, ntree=100,lda=TRUE)
model.rf1
```
